# Image Captioning Project


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import re
```

```python
#Load Data
with open("./rap.txt") as f:
    text= f.read()
```

```python
text
```

```python
#Preproces dATa
text = text.lower().strip()
```

```python
text = re.sub(r"\n+","NEWLINE",text)
```

```python
text[:500]
```

```python
## Vectorize Text
from keras.preprocessing.text import Tokenizer,text_to_word_sequence
```

```python
tokens = text_to_word_sequence(text)
```

```python
tokens
```

```python
len(tokens)
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([tokens])
```

```python
vocab =list( tokenizer.word_index.keys())
```

```python
vocab
```

```python
vocab_size = len(vocab) + 1
```

```python
vocab_size
```

```python
word_to_int=tokenizer.word_index
```

```python
word_to_int
```

```python
int_to_word = {}
for w,i in word_to_int.items():
    int_to_word[i] = w
```

```python
int_to_word
```

```python
# Convert text to vector
sequences = tokenizer.texts_to_sequences([text])[0]
```

```python
print(sequences[:50])
```

```python
int_to_word[2]
```

```python
for val in sequences[:10]:
    print(int_to_word[val])
```

```python
# Generate Training Data
SEQ_LEN = 7
```

```python
X_train = []
Y_train = []
```

```python
for i in range(0,len(sequences)-SEQ_LEN):
    currentx = sequences[i : i + SEQ_LEN]
    currenty = sequences[i+SEQ_LEN]
    X_train.append(currentx)
    Y_train.append(currenty)
```

```python
X_train =np.array(X_train)
Y_train = np.array(Y_train)
```

```python
### One hot encode
from keras.utils import to_categorical
```

```python
Y_train = to_categorical(Y_train,num_classes=vocab_size)
```

```python
Y_train
```

# Create RNN Model
We are going to use LSTM Network which is better than vanilla RNN

```python
from keras.models import Sequential
from keras.layers import Embedding,LSTM,Dense
```

```python
model= Sequential()

# add embedding layer before anything else
embed = Embedding(vocab_size,50,input_length=SEQ_LEN)
model.add(embed)

# first Lstm layer
lstm1 = LSTM(50, return_sequences = True)
model.add(lstm1)

# second lstm
lstm2 = LSTM(25)
model.add(lstm2)

# hidden layer for ANN
model.add(Dense(200,activation= 'relu'))

# output for prediction
model.add(Dense(vocab_size,activation = 'softmax'))
```

```python
model.summary()
```

```python
### Add Optimer
from keras.optimizers import Nadam,Adamax
from keras.losses import categorical_crossentropy

```

```python
opt = Nadam()
```

```python
model.compile(
optimizer = opt,
loss = 'categorical_crossentropy',
metrics = ['accuracy'])
```

```python
history = model.fit(
X_train,Y_train,
validation_split=0.2,
batch_size = 2,
epochs = 2
)
```

```python
## Generate
from keras.preprocessing.sequence import pad_sequences
```

```python
def generate(model,tokenizer,int_to_word,seq_length,initial_words,num_words):
    res = initial_words # list of words since we provide initial word
    while len(res) < num_words:
        vectorized = tokenizer.texts_to_sequences(res)
        vectorized = np.array(vectorized).ravel()
        vectorized = pad_sequences (vectorized,maxlen= seq_length,truncating = 'pre')
#         p = model.predict(vectorized)
#         indices = np.argsort(p)[][:5]
        predictions = model.predict_classes(vectorized)
        next_word = int_to_word[predictions[0]]
        res.append(next_word)
    return res
        
        
        
    
    
```

```python
generate(model,tokenizer,int_to_word,SEQ_LEN,'I am feelin taki '.split(),num_words=20)
```
