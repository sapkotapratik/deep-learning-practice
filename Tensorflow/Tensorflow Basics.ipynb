# Constants ,Variable and Placeholders

```python
# Constant Implementation
import numpy as np
import tensorflow as tf
a = tf.constant(3)
b = tf.constant(5)
c = a+ b
tf.InteractiveSession()
print(c.eval()) # useful in jupyter notebook
with tf.Session() as sess:
    print(sess.run(c))
```

```python
# Variable Implementation
import numpy as np
import tensorflow as tf
a = tf.Variable(0,name='input')
b = tf.constant(1)
mid = tf.add(a,b)
final = tf.assign(a,mid)
var = tf.initialize_all_variables()
with tf.Session() as sess:
    sess.run(var)
    print(sess.run(a))
    for i in range(3):
        sess.run(final)
        print(sess.run(a))
```

```python
# Explain the concept of variable  used in tensorflow
a = tf.Variable(0,name='input')
b = tf.constant(1)

mid_value = tf.add(a,b)
update_value = tf.assign(a,mid_value)

tg = tf.initialize_all_variables()

with tf.Session() as sess:
    sess.run(tg)
    print(sess.run(a))
    
    for i in range(3):
        sess.run(update_value)
        print(sess.run(a))
```

```python
# Placeholder implementation in tensorflow
import numpy as np
import tensorflow as tf

x = tf.placeholder("float",[None,3])
y = x**2
with tf.Session() as sess:
    x_data = [[1,2,3],[4,5,6]]
    result = sess.run(y,feed_dict= {x:x_data})
    print(result)
```

```python
import tensorflow as tf
with tf.name_scope("My_Graph"):
    with tf.name_scope("Scope_1"):
        a = tf.add(1,2,name = "addition_1")
        b = tf.multiply(a,3,name="multiplying_1")
    with tf.name_scope("Scope_2"):
        c = tf.add(4,5,name="addtion_2")
        d = tf.multiply(c,6,name = "multiplying_2")
with tf.name_scope("Scope_3"):
    e = tf.multiply(4,5)
    f = tf.div(c,6,name = "division")
    
g = tf.add(b,d)
h = tf.multiply(g,f)

with tf.Session() as sess:
    writer = tf.summary.FileWriter("./logs",sess.graph)
    print(sess.run(h))
    writer.close()
```

- Randomly pick 100 values from a normal distribution N(0,1) and plot them after the other

```python
import tensorflow as tf

tf.reset_default_graph()

s_scalar = tf.get_variable(name = 'x_scalar',shape=[],initializer = tf.truncated_normal_initializer(mean = 0,stddev =1))

first_summary = tf.summary.scalar(name = 'First_summary',tensor=s_scalar)

init = tf.global_variables_initializer()

with tf.Session() as sess:
    writer = tf.summary.FileWriter('./log',sess.graph)
    for i in range(100):
        sess.run(init)
        summary = sess.run(first_summary)
        writer.add_summary(summary,i)
```
